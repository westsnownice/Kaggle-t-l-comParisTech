```python
import pandas as pd
import numpy as np

df = pd.read_csv('train.csv')
df = df.iloc[:,0].str.split(';', expand=True)
#print df.head(10)
df.rename(columns={0:'age',1:'job',2:'marital',3:'education',4:'default',5:'housing',6:'loan',7:'contact',8:'month',
                   9:'day_of_week',10:'duration',11:'campaign',12:'pdays',13:'previous',14:'poutcome',15:'emp.var.rate',
                   16:'cons.price.idx',17:'cons.conf.idx',18:'nr.euribor3m',19:'nr.employed',20:'y'}, inplace = True)
```

```python
df.head(10)
```

```python
df = df.drop(['month'], axis=1)
df.info()
```

```python
df['marital'].unique()
df['Marital'] = df['marital'].map({'"divorced"':1,'"married"':2,'"single"':3,'"unknown"':4}).astype(int)
df = df.drop(['marital'], axis=1)

df['job'].unique()
df['Job'] = df['job'].map({'"admin."':1,'"blue-collar"':2,'"entrepreneur"':3,'"housemaid"':4,'"management"':5,'"retired"':6,'"self-employed"':7,'"services"':8,'"student"':9,'"technician"':10,'"unemployed"':11,'"unknown"':12}).astype(int)
df = df.drop(['job'], axis=1)

df['contact'].unique()
df[('Contact')] = df['contact'].map({'"cellular"':1,'"telephone"':2}).astype(int)
df = df.drop(['contact'], axis=1)

df['education'].unique()
df['Education'] = df['education'].map({'"basic.9y"':1, '"high.school"':2, '"university.degree"':3,
       '"professional.course"':4, '"basic.6y"':5, '"basic.4y"':6, '"unknown"':7}).astype(int)
#df['Gender'] = df['Sex'].map({'female': 0, 'male':1}).astype(int)
df = df.drop(['education'], axis=1)

#5 - default: has credit in default? (categorical: 'no','yes','unknown')
df['default'].unique()
df['Default'] = df['default'].map({'"no"':0, '"yes"':1, '"unknown"':2}).astype(int)
df = df.drop(['default'], axis=1)

#housing: has housing loan? (categorical: 'no','yes','unknown')
df['housing'].unique()
df['Housing'] = df['housing'].map({'"no"':0, '"yes"':1, '"unknown"':2}).astype(int)
df = df.drop(['housing'], axis=1)

#loan: has personal loan? (categorical: 'no','yes','unknown')
df['loan'].unique()
df['Loan'] = df['loan'].map({'"no"':0, '"yes"':1, '"unknown"':2}).astype(int)
df = df.drop(['loan'], axis=1)

df['day_of_week'].unique()
df['Day_of_week'] = df['day_of_week'].map({'"mon"':1,'"tue"':2,'"wed"':3,'"thu"':4,'"fri"':5}).astype(int)
df = df.drop(['day_of_week'], axis=1)

df['poutcome'].unique()
df['Poutcome'] = df['poutcome'].map({'"failure"':1,'"nonexistent"':2,'"success"':3}).astype(int)

df = df.drop(['poutcome'], axis=1)
```
